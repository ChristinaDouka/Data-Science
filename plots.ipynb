```python
import numpy as np
import matplotlib.pyplot as plt
```

Suppose you have a 1D array, you plot the position of its values on the x axis, they are so dense that you can't tell the spatial distribution, you use a 1D histogram to show the distribution by count of boxes along the x axis. Problem solved.

Then you have two 1D arrays, a list of 2D dots in (x, y) axes. You plot their positions on the x-y plane, again they are so dense and overlap with each other. You want to view the distribution better by count of boxes in the plane, so you try a 2D diagram. Problem solved.

```python
#%matplotlib inline
# prepare 2D random dots centered at (0, 0)
n = 100000
x = np.random.randn(n)
y = x + np.random.randn(n)

# plot data
fig1 = plt.figure()
plt.plot(x,y,'.r')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# plot 2D histogram using pcolor
fig2 = plt.figure()
plt.hist2d(x, y, bins=100)
plt.xlabel('x')
plt.ylabel('y')
cbar = plt.colorbar()
cbar.ax.set_ylabel('Counts')
```

```python
img = plt.imread('testpic.jpg') # Load the image into an array: img

print(img.shape) # Print the shape of the image

plt.imshow(img) # Display the image
plt.axis('off') # Hide the axes
#plt.show()
```

```python
# Compute the sum of the red, green and blue channels: intensity
intensity = img.sum(axis=2)
print(intensity)

# Display the intensity with a colormap of 'gray'
plt.imshow(intensity, cmap='gray')

plt.colorbar() # Add a colorbar
plt.axis('off')
plt.show()
```

```python
#The ratio of the displayed width to height is known = image aspect
#the range used to label the x- and y-axes is known = image extent
#The default aspect value of 'auto' keeps the pixels square and the extents are automatically computed from the shape of the array if not specified otherwise.

# Load the image into an array: img
img = plt.imread('testpic.jpg')

# Specify the extent and aspect ratio of the top left subplot
plt.subplot(2,2,1)
plt.title('extent=(-1,1,-1,1),\naspect=0.5') 
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=0.5)

# Specify the extent and aspect ratio of the top right subplot
plt.subplot(2,2,2)
plt.title('extent=(-1,1,-1,1),\naspect=1')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=1)

# Specify the extent and aspect ratio of the bottom left subplot
plt.subplot(2,2,3)
plt.title('extent=(-1,1,-1,1),\naspect=2')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=2)

# Specify the extent and aspect ratio of the bottom right subplot
plt.subplot(2,2,4)
plt.title('extent=(-2,2,-1,1),\naspect=2')
plt.xticks([-2,-1,0,1,2])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-2,2,-1,1), aspect=2)

# Improve spacing and display the figure
plt.tight_layout()
plt.show()
```

```python
image = plt.imread('testpic.jpg') # Load the image into an array: image

pmin, pmax = image.min(), image.max() # Extract minimum and maximum values from the image: pmin, pmax
print("The smallest & largest pixel intensities are %d & %d." % (pmin, pmax))
plt.subplot(2,1,1)
plt.imshow(image)

# Rescale the pixels: rescaled_image
rescaled_image = 256*(image-pmin)/(pmax-pmin)
print(rescaled_image)
plt.subplot(2,1,2)
plt.imshow(rescaled_image)

print("The rescaled smallest & largest pixel intensities are %.1f & %.1f." % 
      (rescaled_image.min(), rescaled_image.max()))
```

```python
#example online: https://medium.com/@neuralnets/data-visualization-with-python-and-seaborn-part-4-9b0119887130
# Import plotting modules
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

tips = sns.load_dataset('tips') #load the example dataset tips
print(tips.head())
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips) # Plot a linear regression between 'weight' and 'hp'
plt.show()
```

```python
#In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding. 
#Adding `hue="smoker" tells seaborn you want to colour the data points for smoker and non-smoker differently.
sns.lmplot(x='total_bill',y='tip',hue='smoker', data=tips) # Plot a linear regression between 'weight' and 'hp'
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker', markers=['o','x'],palette='Set1', data=tips) # Plot a linear regression between 'weight' and 'hp'
plt.show()
```

```python
#Often, you don't just want to see the regression itself but also see the residuals to get a better idea how well the regression captured the data. Seaborn provides sns.residplot() for that purpose, visualizing how far datapoints diverge from the regression line.
sns.residplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
plt.scatter(tips['total_bill'], tips['tip'], label='data', color='red', marker='o')

sns.regplot(x='total_bill', y='tip', data=tips,color='blue', scatter=None, label='order 1', order=1)

sns.regplot(x='total_bill', y='tip', data=tips, color='green', scatter=None, label='order 2',order=2)

plt.legend('this is a legend', loc="upper right")
plt.show()
```

```python
tips = sns.load_dataset('tips') #load the example dataset tips
print(tips.head())

sns.lmplot(x='total_bill',y='tip',data=tips, row='sex', scatter_kws={"color": "black"}, line_kws={"color": "pink"}) # Plot linear regressions between 'total_bill' and 'tips' grouped row-wise by 'sex'

plt.show()
```

```python
import pandas as pd
auto = pd.read_csv('seaborn-data-master/mpg.csv')
print(auto.head())

#Regressions are useful to understand relationships between two continuous variables. Often we want to explore how the distribution of a single continuous variable is affected by a second categorical variable. Seaborn provides a variety of plot types to perform these types of comparisons between univariate distributions.
#The strip plot is one way of visualizing this kind of data. It plots the distribution of variables for each category as individual datapoints. For vertical strip plots (the default), distributions of continuous values are laid out parallel to the y-axis and the distinct categories are spaced out along the x-axis.

plt.subplot(2,1,1)
sns.stripplot(auto['cylinders'], auto['hp'], data=auto)

# Make the strip plot again using jitter and a smaller point size
plt.subplot(2,1,2)
sns.stripplot(x='cylinders', y='hp', data=auto,size=3, jitter=True)

plt.show()
```

```python
#As you have seen, a strip plot can be visually crowded even with jitter applied and smaller point sizes. 
#An alternative is provided by the swarm plot (sns.swarmplot()), which is very similar but spreads out the points to avoid overlap and provides a better visual overview of the data.

# Generate a swarm plot of 'hp' grouped horizontally by 'cyl'  
plt.subplot(2,1,1)
sns.swarmplot(x='cylinders',y='hp',data=auto)

# Generate a swarm plot of 'hp' grouped vertically by 'cyl' with a hue of 'origin'
plt.subplot(2,1,2)
sns.swarmplot(x='hp',y='cylinders',data=auto, hue='origin', orient='h')

# Display the plot
plt.show()
```

```python
#Both strip and swarm plots visualize all the datapoints. For large datasets, this can result in significant overplotting. Therefore, it is often useful to use plot types which reduce a dataset to more descriptive statistics and provide a good summary of the data. Box and whisker plots are a classic way of summarizing univariate distributions but seaborn provides a more sophisticated extension of the standard box plot, called a violin plot.

```
